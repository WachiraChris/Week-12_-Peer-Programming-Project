Plot the Image

```python
image_A = mnist.iloc[0, :-1].astype('float')

image_array = np.asarray(image_A)
image = image_array.reshape((28, 28))

plt.imshow(image, cmap ='gray')
plt.axis('off');
plt.show()
```

```python
from scipy.ndimage.interpolation import shift
```

Shift Image Up

```python
def shift_image_up(image, shift_pixels = 1):
    shift_vector = [shift_pixels, 0]
    shifted_image = shift(image, shift_vector, cval=0)
    return shifted_image
image_shifted = shift_image_up(image, shift_pixels=1)

fig,axes = plt.subplots(nrows=1, ncols=2, figsize=(10,6))
axes[0].imshow(image, cmap='gray')
axes[0].set_title('Unshifted Image')
axes[1].imshow(image, cmap='gray')
axes[1].set_title('Shifted Image')


plt.show()
```
